```python
import numpy as np
```

```python
X= np.asarray([[1,-5],
               [1,-3],
               [1,0],
               [1,2],
               [1,1]])
```

```python
X
```

```python
Y = np.asarray([-2,1,4,6,3])
```

```python
Beta_hat = np.linalg.inv(np.transpose(X) @ X) @ np.transpose(X) @ Y 
```
